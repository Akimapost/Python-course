Design a simulation where you toss two coins simultaneously.
Record the possible outcomes (heads or tails for each coin) and run the simulation for multiple trials (e.g., 100, 500, or 1000 tosses).
After collecting the data:

1. Calculate the probabilities of each outcome (HH, HT, TH, TT).
2. Visualize the results using a bar chart or pie chart to represent the frequencies and probabilities of each outcome.
3. Analyze whether the results align with the expected theoretical probabilities.

```python
import numpy as np
import matplotlib.pyplot as plt

coin = ['Heads', 'Tails']
number_of_throws = 1000
tosses_coin1 = np.random.choice(coin, number_of_throws)
tosses_coin2 = np.random.choice(coin, number_of_throws)
```

```python
# Number of each outcome (HH, HT, TH, TT)
nr_hh = np.sum((tosses_coin1 == 'Heads') & (tosses_coin2 == 'Heads'))
nr_ht = np.sum((tosses_coin1 == 'Heads') & (tosses_coin2 == 'Tails'))
nr_th = np.sum((tosses_coin1 == 'Tails') & (tosses_coin2 == 'Heads'))
nr_tt = np.sum((tosses_coin1 == 'Tails') & (tosses_coin2 == 'Tails'))

# Store the counts in a dictionary
outcome_counts = {'HH': nr_hh, 'HT': nr_ht, 'TH': nr_th, 'TT': nr_tt}
# Display the results
print("Outcome counts:", outcome_counts)
```

```python
# The probabilities of each outcome (HH, HT, TH, TT)
prob_hh = nr_hh / number_of_throws
prob_ht = nr_ht / number_of_throws
prob_th = nr_th / number_of_throws
prob_tt = nr_tt / number_of_throws

# Store the simulated  probabilities
simulated_probabilities = [prob_hh, prob_ht, prob_th, prob_tt]
```

```python
# Theoretical probabilities (all outcomes are 0.25)
theoretical_probabilities = [0.25, 0.25, 0.25, 0.25]
```

```python
# Create a bar chart to represent the frequencies and probabilities of each outcome
outcomes = ['HH', 'HT', 'TH', 'TT']
x = np.arange(len(outcomes))  # label locations
width = 0.40  # width of the bars

fig, ax = plt.subplots(figsize=(10,10))

# Plot simulated probabilities
bars1 = ax.bar(x - width/2, simulated_probabilities, width, label='Simulated', color='green')

# Plot theoretical probabilities
bars2 = ax.bar(x + width/2, theoretical_probabilities, width, label='Theoretical', color='orange')

# Add labels and title
ax.set_xlabel('Coin Toss Outcomes')
ax.set_ylabel('Probability')
ax.set_title('Simulated vs Theoretical Probabilities for Two Coin Tosses')
ax.set_xticks(x)
ax.set_xticklabels(outcomes)
ax.legend()
plt.show()
```
