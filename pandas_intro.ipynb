```python
import pandas as pd
data_list = [1,2,3,4,5]
print(data_list)
```

```python
###Pandas series
series_from_list = pd.Series(data_list)
print(series_from_list) #prints the indexes of a rows
```

```python
###Pandas DataFrame
dataframe_from_list = pd.DataFrame(data_list)
print(dataframe_from_list) #prints the index of a row and a column
```

```python
###Pandas Series from Dictionary
data_dict = {'A': 10, 'B': 20, 'C': 30, 'D': 40, 'E': 50}
series_from_dict = pd.Series(data_dict)
print(series_from_dict)
```

```python
csv_data = "1,2,3,4,5"
csv_data.split(',') #from string to list of int
```

```python
series_from_list_split = pd.Series(csv_data.split(','))
print(series_from_list_split) #prints the indexes of a rows
```

```python
#DataFrames
#lists are accessed by index and dictionaries with a key
data_list = [
    ['John', 30, 68.5, True, 95.5],
    ['Alice', 25, 63.2, False, 88.0],
    ['Bob', 35, 71.0, True, 76.5],
    ['Eve', 28, 65.8, False, 92.0],
    ['Charlie', 40, 72.3, False, 87.5]
]
column_names = ['Name', 'Age', 'Weight', 'Is Student', 'Score']
pd.DataFrame(data_list, columns=column_names) #nested list become a dataframe
```
