```python
import numpy as np
```

```python
coin = ['Heads', 'Tails'] # 1D array or a list

number_of_throws = 1000 # the size of a rundom list that we want genarate
```

https://numpy.org/doc/stable/reference/random/generated/numpy.random.choice.html numpy.random.choice

```python
tosses = np.random.choice(coin, number_of_throws)
print(f'The result of coin toss simulation for {number_of_throws} throws are: {tosses}')
```

https://numpy.org/doc/stable/reference/generated/numpy.sum.html numpy.sum

```python
# Number of heads
nr_heads = np.sum(tosses == 'Heads')
nr_heads
```

```python
# Number of tails
nr_tails = np.sum(tosses == 'Tails')
nr_tails
```

```python
# Probability of heads
prob_heads = nr_heads / number_of_throws
prob_heads
```

```python
# Probability of tails
prob_tails = nr_tails / number_of_throws
prob_tails
```

# Numpy arrays
https://stackoverflow.com/questions/993984/what-are-the-advantages-of-numpy-over-regular-python-lists

Using NumPy arrays instead of Python lists has several advantages, particularly when working with large datasets or performing mathematical operations. Here are some key benefits of NumPy arrays over Python lists:

###Performance:
NumPy arrays are implemented in C and optimized for performance. They are generally faster than Python lists for numerical operations, especially when dealing with large datasets.

###Memory Efficiency:  
NumPy arrays require less memory compared to Python lists. While a Python list can hold elements of different types, a NumPy array is homogeneous (all elements are of the same type), which allows for better memory utilization.

###Built-in Mathematical Functions:
NumPy provides a wide variety of functions for mathematical operations that can be applied to arrays directly. These functions are optimized for speed and can perform operations on entire arrays or along specific axes.

###Multidimensional Arrays:
NumPy natively supports multidimensional arrays (also known as tensors), which allows for more complex data structures and operations, such as matrix manipulations, without needing to manually nest lists.

###Linear Algebra Support:
NumPy has built-in functions for linear algebra operations, including dot products, matrix inversion, and eigenvalue calculations. This makes it a great choice for scientific computing and data analysis.

###Broadcasting:
NumPy supports broadcasting, which allows for arithmetic operations between arrays of different shapes and sizes without requiring explicit looping. This feature simplifies code and enhances performance.

###Integration with Other Libraries:
Many scientific computing libraries (such as SciPy, Pandas, and Matplotlib) are built on top of NumPy and take advantage of its array structure and performance.

#UNIFORM DISTRIBUTION

```python
coordinates2 = np.random.rand(100000, 2)
coordinates2

x = coordinates2[:, 0] # all the rows from the first column
y = coordinates2[:, 1] # all the rows from the second column

plt.scatter(x, y, color = 'blue')
plt.title('100000 randomly selected coordinates - uniform distribution')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```

#NORMAL DISTRIBUTION

```python
coordinates3 = np.random.randn(100000, 2)
coordinates3

x = coordinates3[:, 0] # all the rows from the first column
y = coordinates3[:, 1] # all the rows from the second column

plt.scatter(x, y, color = 'blue')
plt.title('100000 randomly selected coordinates - normal distribution')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```

```python
# COLORFUL variant
import numpy as np
import matplotlib.pyplot as plt

# Generate 100000 random coordinates
coordinates3 = np.random.randn(100000, 2)

x = coordinates3[:, 0] # X values
y = coordinates3[:, 1] # Y values

# Create a color range based on the y-values or any other function of coordinates
colors = np.arctan2(y, x)  # This will create a circular color gradient

# Create scatter plot with rainbow colors
plt.scatter(x, y, c=colors, cmap='rainbow', alpha=0.7)
plt.title('100000 randomly selected coordinates - normal distribution')
plt.xlabel('X')
plt.ylabel('Y')
plt.colorbar()  # Adds a color bar for reference
plt.grid(True)
plt.show()
```
