Answer the following questions:

How many Pokémons are with 'Type 1' == Water as a % of total?

What is the maximum 'Speed' value? What is the minimum 'Speed' value? What is the difference between max and min 'Speed'?

Filter the DataFrame to include only the Pokémon with 'Speed' >= 80. How many Pokémon meet this criterion?
Display this DataFrame using your preferred visualization method.

(DIFFICULT) Find Pokémon with the longest name (excluding spaces)? What is this pokemons name?

```python
import pandas as pd
import matplotlib.pyplot as plt

pokemon_df = pd.read_csv('/content/Pokemon.csv')

# How many Pokémons are with 'Type 1' == Water as a % of total?
water_pokemon = pokemon_df[pokemon_df['Type 1'] == 'Water']
num_water_pokemon = len(water_pokemon)
total_pokemon = len(pokemon_df)
percentage_water_pokemon = (num_water_pokemon / total_pokemon) * 100
print(f"Percentage of Pokémon with Water Type: {percentage_water_pokemon:.2f}%")

#What is the maximum 'Speed' value? What is the minimum 'Speed' value?
#What is the difference between max and min 'Speed'?
max_speed_value = pokemon_df['Speed'].max()
name_max_speed = pokemon_df['Name'].iloc[pokemon_df['Speed'].idxmax()]
print(f"The {name_max_speed} is the Pokémon with maximum Speed value = {max_speed_value}")

min_speed_value = pokemon_df['Speed'].min()
name_min_speed = pokemon_df['Name'].iloc[pokemon_df['Speed'].idxmin()]
print(f"The {name_min_speed} is the Pokémon with minimum Speed value = {min_speed_value}")
speed_difference = max_speed_value - min_speed_value
print(f'The difference between their speed is: {speed_difference}')

# Filter the DataFrame for Pokémon with Speed >= 80 + visualization
speed_greater_than_80 = pokemon_df[pokemon_df['Speed'] >= 80]
num_pokemons = len(speed_greater_than_80)
print('Number of the Pokémon with Speed >= 80 is: ', num_pokemons)

speed_counts = speed_greater_than_80['Speed'].value_counts().reset_index()
speed_counts.columns = ['Speed', 'Count']

speed_counts.plot(x='Count', y='Speed', kind='bar', color='blue', edgecolor='black')
plt.title('Pokémon count by Speed level >= 80')
plt.xlabel('Count of Pokémon')
plt.ylabel('Speed level')
plt.show()

```
