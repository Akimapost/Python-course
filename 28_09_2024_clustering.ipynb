**Unsupervized traning example (clustering)**

https://scikit-learn.org/dev/api/sklearn.datasets.html

```python
from sklearn.datasets import make_blobs #creating synthetic data
X, y = make_blobs(n_samples=10000, centers=6, cluster_std=0.80)
X[:, 0] #X-coordinate
X[:, 1] #y-coordinate
y # the cluster of the point
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')
```

```python
from sklearn.cluster import KMeans # model
```

```python
kmeans = KMeans(n_clusters=6) # initialize the model
```

```python
kmeans.fit(X) #train model
```

```python
cluster_labels = kmeans.predict(X) # predict
```

```python
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, edgecolors='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, marker='*', c='red') # add cluster centers
```
